```python
import pandas as pd

df = pd.read_csv('copper_mining.csv')
display(df.head())
```

```python
df = df.loc[:, ~df.columns.str.startswith("other")]
display(df.head())
```

```python
df.drop(columns=['sector', 'subsector', 'geometry_ref', 'lat_lon', 'reporting_entity', 'sector_id', 'source_type', 'created_date', 'modified_date'], inplace=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filtrer la mine d'intérêt
mine = "Zimplats Mine"
df_mine = df[df["source_name"] == mine].copy()

# S'assurer que start_time est en datetime
df_mine["start_time"] = pd.to_datetime(df_mine["start_time"], errors="coerce")

# Regrouper par mois
monthly = (
    df_mine
    .groupby(df_mine["start_time"].dt.to_period("M"))["emissions_quantity"]
    .sum()
    .to_timestamp()  # Convertit PeriodIndex → DatetimeIndex pour matplotlib
)

print(monthly)

# --- Visualisation : Scatter Plot ---
plt.figure(figsize=(10, 6))
plt.scatter(monthly.index, monthly.values)

plt.title(f"Émissions mensuelles pour {mine}")
plt.xlabel("Mois")
plt.ylabel("Émissions")
plt.grid(True)
plt.tight_layout()
plt.show()
```
